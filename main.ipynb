## 2. Now write a function that returns the largest element on a list

```python
#your code here
def greatest(lst):
    # Assume the list is not empty
    largest = lst[0]  # Start by assuming the first element is the largest
    
    # Iterate through the rest of the list starting from the second element
    for num in lst[1:]:
        if num > largest:
            largest = num  # Update the largest number if a larger one is found
    
    return largest
```

```python
# This will test your function 
test_greatest(greatest)
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        result = 1
        for i in range(2, n + 1):
            result *= i
        return result

# Example usage:
print(factorial(5))  # Output: 120
print(factorial(0))  # Output: 1
```

```python
#factorial formula
#n! = n * ( n - 1 ) *...*1

# This code defines a function called "factorial" which takes an input "n". The function uses a for loop to iterate through the range of numbers 
# from 1 to n+1. For each number in that range, it multiplies the current value of x by the number in the range. At the end of the loop, 
# the function returns the final value of x, which will be the factorial of the input number "n".

# The Factorial of a positive integer n is the product of all positive integers less than or equal to n. 
# For example, the factorial of 6 (written "6!") is 6 * 5 * 4 * 3 * 2 * 1 = 720.

# So this function takes an input of any positive integer, and returns the factorial of that number.
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(numbers):
    # Step 1: Calculate the mean (average)
    mean = sum(numbers) / len(numbers)
    
    # Step 2: Calculate the squared differences from the mean
    squared_differences = [(x - mean) ** 2 for x in numbers]
    
    # Step 3: Calculate the variance (mean of squared differences)
    variance = sum(squared_differences) / len(numbers)
    
    # Step 4: Calculate the standard deviation (square root of variance)
    standard_deviation = variance ** 0.5
    
    return standard_deviation
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(s):
    # Convert the string to lowercase to handle case insensitivity
    s = s.lower()
    
    # Create a set of all letters in the alphabet
    alphabet = set('abcdefghijklmnopqrstuvwxyz')
    
    # Create a set of letters found in the string (only alphabet characters)
    letters_in_string = set(char for char in s if char.isalpha())
    
    # Check if the letters in the string contain all the letters of the alphabet
    return letters_in_string == alphabet
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(s):
    # Convert the string to a list of words by splitting it manually using commas
    words = []
    current_word = ""
    
    for char in s:
        if char == ',':
            words.append(current_word)
            current_word = ""
        else:
            current_word += char
    
    # Append the last word after finishing the loop
    if current_word:
        words.append(current_word)
    
    # Sort the words alphabetically
    sorted_words = sorted(words)
    
    # Convert the sorted list back to a string with commas separating the words
    result = ""
    for i, word in enumerate(sorted_words):
        if i != 0:
            result += ','
        result += word
    
    return result
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.

```python
import string

def check_pass(password):
    # Check the length of the password
    if len(password) < 8:
        return False
    
    # Initialize flags for each condition
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    
    # Iterate through each character in the password
    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif char in string.punctuation:  # Check if it's a special character
            has_special = True
    
    # Return True only if all conditions are met
    return has_lower and has_upper and has_digit and has_special
```

```python
# This will test your function 
test_pass(check_pass)
```
